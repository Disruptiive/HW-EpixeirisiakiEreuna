```python
import gurobipy as gp
from gurobipy import GRB
```

```python
prev_color = ["1", "2","3","4","5"]

next_color = ["1", "2","3","4","5"]

n1 = len(prev_color)

n2 = len(next_color)

demand = [500,500,500,500,500]

fixedCosts = [40,35,45,32,50]

transCosts = [[0,11,7,13,11],
              [5,0,13,15,15],
              [13,15,0,23,11],
              [9,13,5,0,3],
              [3,7,7,7,0]]

```

```python
from itertools import chain, combinations
colors = list(range(n1))
xd = list(chain.from_iterable(combinations(colors, r) for r in range(len(colors)+1)))

banned_sets = xd[1:len(xd)-1]
```

```python
m = gp.Model("Colors")

x = m.addVars(n1,n2,vtype=GRB.INTEGER,name='Assign')

m.setObjective(gp.quicksum((fixedCosts[j]+transCosts[i][j]+fixedCosts[i])*x[i,j] for j in range(n1) for i in range(n2)), GRB.MINIMIZE)

for i in range(n1):
    for j in range(n2):
        if i == j:
            m.addConstr(x[i,j] == 0)
    #m.addConstr(gp.quicksum(x[i,j] for j in range(n2))>=demand[i])
    m.addConstr(gp.quicksum(x[i,j] for j in range(n2)) == 1)
    m.addConstr(gp.quicksum(x[j,i] for j in range(n2)) == 1)
    
for S in banned_sets:
    m.addConstr(gp.quicksum(x[i,j] for j in S for i in S)<=len(S)-1,"tour_elim_%d" % sum(S))


```

```python
m.write('xd.lp')
```

```python
m.optimize()
```
