```python
import gurobipy as gp
from gurobipy import GRB
```

```python
orders = ["1", "2","3"]

machines = ["1", "2","3"]

n_orders = len(orders)

n_machines = len(machines)

duration = [[45,20,12],
            [0,10,17],
            [10,34,28]]
```

```python
m = gp.Model("Wallpaper")
start_time = m.addVars(n_orders,n_machines,ub=200,vtype=GRB.INTEGER,name='Staring Time')
end_time =  m.addVars(n_orders,n_machines,ub=200,vtype=GRB.INTEGER,name='Ending Time')
max_time =  m.addVar(vtype=GRB.INTEGER,name='Max Time')

b = m.addVars(n_orders,n_machines,vtype=GRB.BINARY,name='b')

m.setObjective(max_time, GRB.MINIMIZE)

m.addConstr(max_time == gp.max_(end_time[i,j] for i in range(n_machines) for j in range(n_orders)))

for i in range(n_machines): 
    for j in range(n_orders):
        m.addConstr(end_time[i,j] == start_time[i,j] + duration[i][j])
        

m.addConstr(end_time[0,0] <= start_time[2,0])

m.addConstr(end_time[1,1] <= start_time[0,1])
m.addConstr(end_time[0,1] <= start_time[2,1])

m.addConstr(end_time[2,2] <= start_time[0,2])
m.addConstr(end_time[0,2] <= start_time[1,2])

m.addConstr(end_time[1,0]  == 0)
m.addConstr(start_time[1,0] == 0)

 
```

```python
lst = [(0,1),(0,2),(1,2)]
eps = 0.0001
M = 200 + eps
for i in range(n_machines): 
    for j,k in enumerate(lst):
        m.addConstr(start_time[i,k[0]] >= start_time[i,k[1]] + eps - M*(1-b[i,j]))
        m.addConstr(start_time[i,k[0]] <= start_time[i,k[1]] +  M*b[i,j])
        m.addConstr((b[i,j] == 1) >> (start_time[i,k[0]] >= end_time[i,k[1]]))
        m.addConstr((b[i,j] == 0) >> (start_time[i,k[1]] >= end_time[i,k[0]]))
```

```python
m.write('xd2.lp')
```

```python
m.optimize()
```
